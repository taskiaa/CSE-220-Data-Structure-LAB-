```python
class MaxHeap:
    def __init__(self, A):
        self.B = [0] * A
        self.C = 0
        self.D = A

    def insert(self, E):
        if self.C == self.D:
            return
        self.C = self.C + 1
        self.B[self.C - 1] = E
        self._swim(self.C - 1)

    def _swim(self, F):
        while F > 0 and self.B[F] > self.B[(F - 1) // 2]:
            self.B[F], self.B[(F - 1) // 2] = self.B[(F - 1) // 2], self.B[F]
            F = (F - 1) // 2

    def extract_max(self):
        if self.C == 0:
            return None
        G = self.B[0]
        self.B[0] = self.B[self.C - 1]
        self.C = self.C - 1
        self._sink(0)
        return G

    def _sink(self, H):
        while 2 * H + 1 < self.C:
            I = 2 * H + 1
            J = 2 * H + 2
            K = H
            if I < self.C and self.B[I] > self.B[K]:
                K = I
            if J < self.C and self.B[J] > self.B[K]:
                K = J
            if K == H:
                break
            self.B[H], self.B[K] = self.B[K], self.B[H]
            H = K

def find_top_k_largest(A, B):
    C = MaxHeap(len(A))
    for D in A:
        C.insert(D)
    E = [0] * B
    for F in range(B):
        E[F] = C.extract_max()
    return E


A = list(map(int, input("Enter the elements of the array separated by spaces: ").split()))
B = int(input("Enter the number of largest elements to find (k): "))


C = find_top_k_largest(A, B)
print("The top", B, "largest elements are:", C)
```
