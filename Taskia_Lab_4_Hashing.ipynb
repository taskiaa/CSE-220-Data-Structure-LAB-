```python
class Node:
  def __init__(self, value=None, next = None):
    self.value = value
    self.next = next

class HashTable:
  def __init__(self, length):
    n = Node()
    self.ht = [n] * length
    self.length = length

  def show(self):
    count = 0
    for i in self.ht:
      temp = i
      print(count, end=" ")
      while temp!=None:
        print (temp.value, end="-->")
        temp = temp.next
      count += 1
      print()


  #Do it by yourself
  def __hash_function(self, key):
    #TO DO
    final= 0
    i = 0
    if len(key) % 2 == 0:
        while i < len(key):
            final += ord(key[i])
            i += 2
    else:
        i = 1
        while i < len(key):
            final += ord(key[i])
            i += 2
    return final % self.length



  #Do it by yourself
  def insert(self, key, value):
    #TO DO
    things = (key, value)
    n_node = Node(things)
    location = self.__hash_function(key)

    if self.ht[location].value is None:
        self.ht[location] = n_node
    else:
        head = self.ht[location]

        if value > head.value[1]:
            n_node.next = head
            self.ht[location] = n_node
        else:
            present = head
            while present.next is not None and present.next.value[1] >= value:
                present = present.next


            # Insert the new node
            n_node.next = present.next
            present.next = n_node
```

```python
class Node_pair:
  def __init__(self, key, value, next = None):
    self.key, self.value, self.next = key, value, next


class Hashtable:
  def __init__(self, size):
    self.ht = [None]*size


  def insert(self, s):
    hashed_index = self.__hash_function(s[0])
    pair = Node_pair(s[0], s[1])
    if self.ht[hashed_index] == None:
      self.ht[hashed_index] = pair
    else:
      pair.next = self.ht[hashed_index]
      self.ht[hashed_index] = pair


  def create_from_array(self, arr):
    for i in arr:
      self.insert(i)

  def print_hashtable(self):
    idx = 0
    for i in self.ht:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'({head.key}, {head.value})', end = '-->')
        head = head.next
      print('None')
      idx += 1


  #Do it by yourself
  def __hash_function(self, key):
    #TO DO
    hash= key +3
    return hash % len(self.ht)




  #Do it by yourself
  def remove(self, key):
    #TO DO

    location = self.__hash_function(key)

    if self.ht[location] == None:
        return self.ht

    present = self.ht[location]

    # If the key is at the head of the chain
    if present.key == key:
        self.ht[location] = present.next
        return self.ht

    # Using a for loop with iteration on the chain
    for x in range(len(self.ht)):
        n_node = present.next
        if n_node == None:
            break
        if n_node.key == key:
            present.next = n_node.next
            return self.ht
        present = n_node

    return self.ht



```
